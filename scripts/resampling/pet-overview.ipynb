# PET数据质量预览

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import matplotlib
import matplotlib as mpl
%config InlineBackend.figure_format = 'svg'
```

```python
qc_frame = pd.read_csv('../datas/MODIS/MODIS16A2GF/MOD16A2GF-006-ET-QC-500m-Statistics-QA.csv')
qc_frame
```

```python
qc_frame[qc_frame.Date.str.contains('2000')]
```

```python
qc_frame[qc_frame.Date.str.contains('2000')].sum().iloc[3:]
```

```python
year_qc = np.array([qc_frame[qc_frame.Date.str.contains(str(i))].sum().iloc[3:].to_list() for i in range(2000,2016)])
years = [i for i in range(2000,2016)]
quality_level = np.array([0,8,16,32,40,48,97,105,113,157])
```

```python
fig, ax = plt.subplots(figsize=[6,6])
im = ax.imshow(year_qc,  interpolation='nearest',cmap="jet")
ax.set_xticks(np.arange(len(quality_level)),labels=quality_level)
ax.set_yticks(np.arange(len(years)),labels=years)
plt.setp(ax.get_xticklabels(), rotation=45, ha="right",
         rotation_mode="anchor")

fig.colorbar(im, ax=ax,use_gridspec=True,format='{x:.0e}',label="Nunber of Rasters")

ax.set_title("MODIS Data Quality Overview")
ax.set_xlabel("Quality Level (Lower is better)")
ax.set_ylabel("Year")
fig.show()
```
